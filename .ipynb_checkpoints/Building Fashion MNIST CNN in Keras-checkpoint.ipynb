## Building A CNN Using Keras
#### Dataset Fashion MNIST

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Input,Convolution2D,MaxPooling2D,Flatten,Dense,Dropout
from keras.utils import np_utils
import tensorflow
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv('./fashionmnist/fashion-mnist_train.csv')
X_ = np.array(x)
X = X_[:,1:] # because first column consists of the labels
X = X/255.0
y = X_[:,0]
print(X.shape,y.shape)
```

```python
np.unique(y,return_counts=True) # Well Balanced dataset
```

```python
# Since we know that CNN excepts data only in the form of images, so we will first reshape the data in the form of images
X_train = X.reshape((-1,28,28,1)) 
Y_train = np_utils.to_categorical(y)
#print(Y_train)
print(X_train.shape,Y_train.shape)
```

```python
for i in range(10):
    plt.imshow(X_train[i].reshape(28,28),cmap="gray")
    plt.show()
    
```

## CNN Model

```python
model = Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(Dropout(0.25)) # adding dropout layer. 0.25 means that randomly dropping 25% of the neuraons from the previous layer
model.add(MaxPooling2D(2,2))
model.add(Convolution2D(32,(5,5),activation='relu'))
model.add(Convolution2D(8,(5,5),activation='relu'))
model.add(Flatten())
model.add(Dense(10,activation='softmax')) # this layer gives probabilities for the 10 classes and applies softmax on all the 10 probabilities
model.summary()
```

```python
 model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
hist=  model.fit(X_train,Y_train,epochs=20,shuffle=True,batch_size=256,validation_split=0.20) # 0.20 means take 20% of the total data as validation data
```

```python
plt.figure(0)
plt.plot(hist.history['loss'],'g')
plt.plot(hist.history['val_loss'],'b')

plt.plot(hist.history['accuracy'],'r')
plt.plot(hist.history['val_accuracy'],'black')
plt.show()
```
