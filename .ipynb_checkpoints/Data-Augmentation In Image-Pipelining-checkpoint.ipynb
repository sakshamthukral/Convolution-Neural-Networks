## Data Augmentation and Image Pipeline 
- Real Life Datasets can be large or small
- Image Pipeline is an efficient way to handle large datasets
- Data Augmentation is a technique of augmenting data,which acts as a regularizer and hence reduces overfitting.

```python
import numpy as np
import keras
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os
```

```python
# Applying augmentation using generators
# generators apply a set of transformations on the data
image_augmentation_for_generator = ImageDataGenerator(  
    rescale = 1/255.0,  # scales every image in the range 0 to 1 
    rotation_range = 40, # 40 degrees
    width_shift_range = 0.2, # u should be able to shift your some of the pixels 20% to the left or 20% to the rightt
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.3, # it sometimes happen that the images on which you train ypur model are very small , however while testing u give larger images which thereby results in 
    #wrong outputs, so by passing this argument we are also training our model on a zoomed version of the train dataset
    horizontal_flip = True,
)

# in test generator we will be performing only one transformation ie of rescaling of image in range 0 to 1
test_image_gen = ImageDataGenerator(
    rescale = 1/255.0
)
```

```python
train_gen = image_augmentation_for_generator.flow_from_directory(
    "C:/Users/hp/Datasets/tiny-imagenet-200/train",
    target_size=((224,224)),
    batch_size=128,
    class_mode = 'categorical'
)
test_gen = test_image_gen.flow_from_directory(
    "C:/Users/hp/Datasets/tiny-imagenet-200/test",
    target_size=((224,224)),
    batch_size=128,
    class_mode='categorical'
)
#.flow_from_directory automatically creates a dictionary which maps all diffrent folders of images with different class indices
```

```python
train_gen.class_indices # class_indices is a property of generator which tells us how the folders and the classes are mapped ie which folder is given which class  
```

```python
class_ids = train_gen.class_indices
print(class_ids)
```

```python
 def load_validation_data(target_size,no_of_classes):
        with open("C:/Users/hp/Datasets/tiny-imagenet-200/val/val_annotations.txt") as f:
            lines = f.readlines()
            
            m = len(lines)
            X = np.empty((m,*target_size,3)) # since target size is a tupple so *target_size will unpack those tupples
            # so if you have(224,224) as target_size,then *target size will remove the bracket ie will make 224,224,ie now the batch size will be 
            # batch_size = M x Width=224 x height=224 x 3
            Y = np.empty(m) # similary Y will be an array like this will be having m number of examples inside it
            
            for i,line in enumerate(lines):
                tokens=line.split()
                img_name=tokens[0]
                img_label=tokens[1]
                
                img_url=os.path.join("C:/Users/hp/Datasets/tiny-imagenet-200/val/images",img_name)
                img = image.load_img(img_url)
                img = img.resize(target_size)
                X[i,] = np.array(img,dtype=np.float32)/255.0
                img.close()
                Y[i] = class_ids[img_label]
            return X,keras.utils.to_categorical(Y,no_of_classes)
```

```python
X_val,Y_val = load_validation_data((224,224),200)
```

 ## CNN Sequential Model

```python
from keras.models import Sequential
from keras.layers import Input,Conv2D,MaxPool2D,Flatten,Dense,Dropout
```

```python
model = Sequential()

# 1st Convolution Layer
model.add(Conv2D(filters=96,input_shape=(224,224,3),kernel_size=(11,11),strides=(4,4),padding="valid",activation="relu"))

# Max Pooling
model.add(MaxPool2D(pool_size=(3,3),strides=(2,2),padding="valid"))

# 2nd Convolution Layer
model.add(Conv2D(filters=256,kernel_size=(5,5),strides=(1,1),padding="same",activation="relu"))

# Max Pooling
model.add(MaxPool2D(pool_size=(3,3),strides=(2,2),padding="valid"))

# 3rd Convolution Layer
model.add(Conv2D(filters=384,kernel_size=(3,3),strides=(1,1),padding="same",activation="relu"))

# 4th Convolution layer
model.add(Conv2D(filters=384,kernel_size=(3,3),strides=(1,1),padding="same",activation="relu"))

# 5th Convolution Layer
model.add(Conv2D(filters=384,kernel_size=(3,3),strides=(1,1),padding="same",activation="relu"))

# Max pooling
model.add(MaxPool2D(pool_size=(3,3),strides=(2,2),padding="valid"))

# Passing it to a fully connected layer
model.add(Flatten())

# 1st fully connected layer
model.add(Dense(units=9216,activation="relu"))

# 2nd fully connected layer
model.add(Dense(units=4096,activation="relu"))

# 3rd fully connected layer
model.add(Dense(units=4096,activation="relu"))

# output layer
model.add(Dense(200,activation="softmax")) # as we have two classes
model.summary()
```

### Compile the model and set the optimizers

```python
from keras.optimizers import Adam
```

```python
opt=Adam(lr=0.001)
model.compile(optimizer=opt,loss="categorical_crossentropy",metrics=['accuracy'])
```

### Training Our Model

- Now we need to train our model , we have used two callbacks(ModelCheckpoint and EarlyStopping)

```python
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
checkpoint = ModelCheckpoint("alexnet_1.h5",monitor='val_acc',verbose=1,save_best_only=True,save_weights_only=False,mode='auto',period=1)
early = EarlyStopping(monitor='val_acc',min_delta=0,patience=10,verbose=1,mode='auto')
hist = model.fit_generator(steps_per_epoch=10,generator=train_gen,validation_data=(X_val,Y_val),validation_steps=X_val.shape[0]/32,callbacks=[checkpoint,early])
```
