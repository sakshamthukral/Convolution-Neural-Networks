#### Code Example:
Let's go through one example to understand "Filters" and "Convolution"

When the feature is present in part of an image,the convolution operation between the filter and that part of the image results in a real number with a high value.If the feature is not present the resulting value is low.

![SegmentLocal](Images_Used/working_of_filter.gif "segment")

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img,title='Image'):
    plt.imshow(img,cmap='gray')
    plt.axis("off")
    plt.style.use("seaborn")
    plt.title(title + str(img.shape))
    plt.show()
    
```

```python
img_ = cv2.imread("./Images Used/picture-of-cartoon.jpg") # we know that by default when open cv reads the images , they are in BGR format
img=cv2.resize(img_,(100,100)) # resiszed image
# actual images are rgb images but opencv basically swaps the red and the blue channels
# But if i want to get the actual image i first need to convert it into an RGB image
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImg(img_rgb)
drawImg(img_gray)
```

```python
# Hardcoded filter

# Actually when you build a neural network these filters are learned by the network itself,but here,i am just going to hardcode a filter to see what filters actually do .
# SO there are some popular filters in computer vision , one for detecting boundaries, one for detecting contrasts, one for blurring the image(gaussian blur filter).
# Filter is nothing , but a matrix, which we are going to convolve over the whole image

def convolution(img,img_filter):
    W = img.shape[0] # width of the image
    H = img.shape[1] # height of the image
    
    # assuming a square filter
    F = img_filter.shape[0] #3
    
    new_img = np.zeros((W-F+1,H-F+1)) # new image in which we will be storing the activation values
    
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    # The outer two loops basically iterate over the new_img and the innner two loops iterate over the filter to take element wise product.
                    new_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                
                # The following two checks ensure that we have only valid pixel values or activation values in the new_img:-
                
                # Check 1:its quite possible then when we are multiplying and then adding to the pixel values of the new_img, the pixel value might exceed 255.
                # and as we know that all the value should be in the region 0-255, so prevent this we will put a check as follows:
                if new_img[row][col] >255:
                    new_img[row][col] = 255
                # CHeck 2: its also possible that your output actiavtion value might become negative
                elif new_img[row][col] < 0:
                    new_img[row][col]=0
    return new_img  # Which is basically our activation map
                    
```

```python
#img.shape
```

```python
# 1.blur_filter
blur_filter = np.ones((3,3))/9.0
#print(blur_filter)
output1 = convolution(img_gray,blur_filter)
drawImg(img_gray)
drawImg(output1)
```

```python
# 2.edge_filter(it is one of the most popular filters out there)
edge_filter = np.array([[1,0,-1],
                       [1,0,-1],
                       [1,0,-1]])
output2 = convolution(img_gray,edge_filter)
drawImg(img_gray)
drawImg(output2)
```

### Convolution Layer
- A convolution layer contains various filters,each layer in a Convolution Neural Network is going to contain multiple filters.
- Each filter etracts different kinds of features and gives 1 Activation map
- Multiple activation maps are combined by stacking to form output volume.
- So CNN layer takes input a volume and produces an output volume of different shape.

#### Stride
- Filters can have different size as well as movement
- Stride defines how a filter should move across the image
- Number of pixels we skip each time is called stride
- In our example we used a stride of (1,1) along W and H
- You can also use a stride of (2,2) in that case the output volume will have less W and H

### Padding
- Convolution operation we have seen, reduces "H" and "W" of the original image
- But sometimes we want the output image to have same size as the input image
- So we can achieve this by adding 0 value pixels(neurons) outside the original image
- This is called Padding
<b>Input and Output Sizes after Convolution:</b>(with padding)
<img src="Images_Used/input-outputSizes_after_Convolution2.PNG" height=500 width=500/>

```python
print(img.shape)
drawImg(img)

# Padding to this image
pad_img = np.pad(img,((10,10),(20,20),(0,0)),'constant',constant_values=200) # (10,10) is the padding across the rows ie the 1st axis,(20,20) is the padding across columns ie the 2nd axis and (0,0) is the padding across the channels ie the 3rd axis
drawImg(pad_img)                                                           # 10,10 represents that the padding of 10 before the axis and a padding of 10 after the axis, same of all other (a,b) paddings, a represents the padding before axis and b represnts the padding after axis
```
